```python
# ë°˜ë ¤ë™ë¬¼ ì •ë³´ë¥¼ ë‹´ëŠ” í´ë˜ìŠ¤ ì •ì˜
class Pet:
    # ìƒì„±ì ë©”ì„œë“œ : ê°ì²´ê°€ ìƒì„±ë  ë•Œ ìë™ìœ¼ë¡œ í˜¸ì¶œë¨
    def __init__(self, name: str, species: str, age: int):
        # selfëŠ” ì¸ìŠ¤í„´ìŠ¤ ìì‹ ì„ ì˜ë¯¸
        # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„ ì–¸ : ê°ì²´ë§ˆë‹¤ ê°œë³„ì ìœ¼ë¡œ ê°’ì„ ê°€ì§
        self.name = name
        self.species = species
        self.age = age

    # __str__ë©”ì„œë“œ: ê°ì²´ë¥¼ printí•  ë•Œ ì–´ë–¤ ë¬¸ìì—´ë¡œ ë³´ì—¬ì¤„ì§€ ì •ì˜
    def __str__(self):
        # f=string: ë¬¸ìì—´ ì•ˆì— ë³€ìˆ˜ ì‚½ì… ê°€ëŠ¥
        return f"{self.name}({self.species}) {self.age}ì‚´"

    # ê°ì²´ì˜ ì •ë³´ë¥¼ ìˆ˜ì •í•˜ëŠ” ë©”ì„œë“œë“œ
    def update_info(self, name=None, species=None, age=None):
        # ê¸°ë³¸ê°’ None : ì¸ìê°€ ìƒëµë  ìˆ˜ ìˆìŒ
        if name:    # ë¹ˆ ë¬¸ìì—´ì´ ì•„ë‹ˆë¼ë©´ (ë¬¸ìì—´ì€ ë¹„ì–´ ìˆìœ¼ë©´ Falseë¡œ í‰ê°€ë¨)
            self.name = name
        if species:
            self.species = species
        # ìˆ«ìí˜• ageëŠ” 0ë„ Falseì´ë¯€ë¡œ Noneê³¼ êµ¬ë¶„í•´ì„œ ì²´í¬í•´ì•¼ í•¨
        if age is not None:
            self.age = age

# í•¨ìˆ˜ ì •ì˜: ë¦¬ìŠ¤íŠ¸ì— ìƒˆ ë°˜ë ¤ë™ë¬¼ì„ ì¶”ê°€
def add_pet(pet_list: list):
    # input(): ì‚¬ìš©ìë¡œë¶€í„° ë¬¸ìì—´ ì…ë ¥ ë°›ê¸°ê¸°
    name = input("ì´ë¦„: ")
    species = input("ì¢…ë¥˜: ")
    age = input("ë‚˜ì´: ")   # ë¬¸ìì—´ë¡œ ì…ë ¥ë˜ë¯€ë¡œ ì‹¤ì œ ì‚¬ìš© ì‹œ intë¡œ ë°”ê¾¸ëŠ” ê²ƒì´ ë°”ëŒì§í•¨
    
    # Pet í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„°ìŠ¤(ê°ì²´) ìƒì„±
    pet = Pet(name, species, age)
    
    # ë¦¬ìŠ¤íŠ¸ì— ê°ì²´ ì¶”ê°€ (.append())
    pet_list.append(pet)
    
    print("ë°˜ë ¤ë™ë¬¼ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")

# í•¨ìˆ˜ ì •ì˜: ë“±ë¡ëœ ë°˜ë ¤ë™ë¬¼ ëª©ë¡ ì¶œë ¥
def show_pet(pet_list: list):
    # ë¹ˆ ë¦¬ìŠ¤íŠ¸ ì²´í¬ (Noneì´ ì•„ë‹ˆë¼ ë¹ˆ ë¦¬ìŠ¤íŠ¸ì¸ ê²½ìš° í™•ì¸)
    if pet_list is None:
        print("ë°˜ë ¤ë™ë¬¼ì´ ì—†ìŠµë‹ˆë‹¤.")
    # enumerate(): ì¸ë±ìŠ¤ì™€ ìš”ì†Œë¥¼ í•¨ê»˜ ë°˜í™˜
    for i, pet in enumerate(pet_list):
        print(f"{i+1}. {pet}")

# í•¨ìˆ˜ ì •ì˜: ë°˜ë ¤ë™ë¬¼ ì •ë³´ ìˆ˜ì •
def update_pet(pet_list: list):
    show_pet(pet_list)  # í˜„ì¬ ëª©ë¡ ì¶œë ¥
    
    # ì¸ë±ìŠ¤ë¥¼ ì…ë ¥ë°›ì•„ ë¦¬ìŠ¤íŠ¸ì—ì„œ í•­ëª© ì„ íƒ (ì¸ë±ìŠ¤ëŠ” 0ë¶€í„° ì‹œì‘í•˜ë¯€ë¡œ -1)
    index = int(input("ìˆ˜ì •í•  ë°˜ë ¤ë™ë¬¼ ë²ˆí˜¸: ")) - 1
    
    # ì¸ë±ìŠ¤ ìœ íš¨ì„± ê²€ì‚¬
    if 0 <= index < len(pet_list):
        name = input("ìƒˆ ì´ë¦„ (ê±´ë„ˆë›°ë ¤ë©´ Enter): ")
        species = input("ìƒˆ ì¢…ë¥˜(ê±´ë„ˆë›°ë ¤ë©´ Enter): ")
        age = input("ìƒˆ ë‚˜ì´(ê±´ë„ˆë›°ë ¤ë©´ Enter): ")

        # or None: ë¹ˆ ë¬¸ìì—´("")ì¼ ê²½ìš° Noneìœ¼ë¡œ ëŒ€ì²´í•˜ì—¬ update-infoì—ì„œ ë¬´ì‹œë˜ë„ë¡ í•¨
        pet_list[index].update_info(name or None, species or None, age or None)
        print("ì •ë³´ê°€ ìˆ˜ì •ë˜ì—ˆìŠµë‹ˆë‹¤.")
    else:
        print("ì˜ëª»ëœ ë²ˆí˜¸ì…ë‹ˆë‹¤.")

# í•¨ìˆ˜ ì •ì˜: ë°˜ë ¤ë™ë¬¼ ì‚­ì œ
def delete_pet(pet_list: list):
    show_pet(pet_list)  
    index = int(input("ì‚­ì œí•  ë°˜ë ¤ë™ë¬¼ ë²ˆí˜¸: ")) - 1
    if 0 <= index < len(pet_list):

        # ë¦¬ìŠ¤íŠ¸ì—ì„œ ì¸ë±ìŠ¤ë¡œ ì‚­ì œ (.pop()ì€ ì‚­ì œí•˜ë©´ì„œ ê°’ì„ ë°˜í™˜í•¨)
        deleted = pet_list.pop(index)
        print(f"{deleted.name}ì´(ê°€) ì‚­ì œë˜ì—ˆìŠµë‹ˆë‹¤.")
    else:
        print("ì˜ëª»ëœ ë²ˆí˜¸ì…ë‹ˆë‹¤.")

# ë©”ì¸ í•¨ìˆ˜: ì „ì²´ í”„ë¡œê·¸ë¨ ì‹¤í–‰ì„ ì œì–´í•˜ëŠ” ì—­í• 
def main():
    pet_list = []   # ë°˜ë ¤ë™ë¬¼ ê°ì²´ë“¤ì„ ì €ì¥í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸

    while True: # ë¬´í•œ ë°˜ë³µë¬¸ -> ì‚¬ìš©ì ì¢…ë£Œ ì‹œê¹Œì§€ ì‹¤í–‰
        print("\n[ ë°˜ë ¤ë™ë¬¼ ê´€ë¦¬ ì‹œìŠ¤í…œ ]")
        print("1. ì¶”ê°€")
        print("2. ëª©ë¡ ë³´ê¸°")
        print("3. ì •ë³´ ìˆ˜ì •")
        print("4. ì‚­ì œ")
        print("5. ì¢…ë£Œ")

        # ì‚¬ìš©ì ì„ íƒ ì…ë ¥
        choice = input("ì„ íƒ >> ")

        # ì¡°ê±´ë¬¸ if-elif-elseë¡œ ë©”ë‰´ ê¸°ëŠ¥ ë¶„ë¦¬
        if choice == "1":
            add_pet(pet_list)
        elif choice == "2":
            show_pet(pet_list)
        elif choice == "3":
            update_pet(pet_list)
        elif choice == "4":
            delete_pet(pet_list)
        elif choice == "5":
            print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break   # while ë£¨í”„ ì¢…ë£Œ -> í”„ë¡œê·¸ë¨ ì¢…ë£Œ

# ì‹¤í–‰ ì¡°ê±´: ì´ íŒŒì¼ì„ ì§ì ‘ ì‹¤í–‰í–ˆì„ ë•Œë§Œ main() ì‹¤í–‰
# ë‹¤ë¥¸ íŒŒì¼ì—ì„œ import ë˜ëŠ” ê²½ìš°ì—ëŠ” ì‹¤í–‰ë˜ì§€ ì•ŠìŒ
if __name__ == "__main__":
    main()
```

🔢 4단계: 리스트
문제 4. 5개의 정수 입력받아 평균 구하기

입력: 10 20 30 40 50
출력: 평균은 30.0입니다.

```python

# 여러 개를 받아야 하면 리스트를 떠올리자올리자
numbers = []

for i in range(1, 6):
    num = int(input(f"{i}번째 숫자를 입력하세요:"))
    numbers.append(num)

total_sum = sum(numbers)
total_avg = total_sum / 5

print(f"평균은 {total_avg}입니다.")
```

🧪 10단계: 간단한 숫자 맞추기 게임
문제 10. 숫자 맞추기 게임
1부터 10 사이의 숫자 중 하나를 랜덤으로 고르고, 사용자가 맞출 때까지 반복하세요.
힌트: random 모듈 사용

```python
import random

rd_num = random.randint(1, 10)
print(rd_num)

while True:
    guess = int(input("1~10 사이의 숫자 중 맞춰보세요:"))
    if guess > rd_num:
        print("너무 커요!")
    elif guess < rd_num:
        print("너무 작아요!")
    else:
        print("정답입니다!")
        break
```

🧾 1. 단어 개수 세기 (문자열 처리 + 딕셔너리)
사용자로부터 문장을 입력받아 각 단어가 몇 번 나왔는지 세어 출력하세요.

입력 예시:I love python because python is fun
출력 예시:
I: 1  
love: 1  
python: 2  
because: 1  
is: 1  
fun: 1


```python
str = input("문장을 입력하세요:")
lst = str.split()

word_count = {}
for word in lst:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
        
for word, count in word_count.items():
    print(f"{word}: {count}")
    
```

🧑‍🍳 2. 음식 주문 시스템 (조건문 + 반복문 + 딕셔너리)
간단한 메뉴와 가격표를 만들고, 사용자가 음식 이름을 입력하면 가격을 출력하세요.
종료하려면 "종료"를 입력하세요.

예시 메뉴:
김밥: 3000  
라면: 3500  
떡볶이: 4000

```python
menu = {"김밥": 3000, "라면": 3500, "떡볶이": 4000}

while True:
   order = input("음식을 주문하세요 (종료하려면 '종료' 입력)")
   if order == '종료':
       print("주문을 종료합니다.")
       break
   elif order in menu:
       print(f"{order}의 가격은 {menu[order]}입니다.")
   else:
       print("주문하신 메뉴가 없습니다.")
```

Matplot 연습문제

✅ 1단계: 기본 선 그래프 그리기
목표: matplotlib의 기본 사용법 익히기

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title('Line Grapth')
plt.xlabel("X")
plt.ylabel('Y')
plt.show()
```

✅ 2단계: 여러 개의 선 그리기
목표: 여러 시리즈를 하나의 그래프에 표시

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.plot(x, y1, label="제곱곱")
plt.plot(x, y2, label="역제곱")
plt.title('MutiLine Graph')
plt.legend()
plt.show()
```

✅ 3단계: 막대 그래프 (Bar Chart)
목표: 카테고리 데이터를 시각화

```python
labels = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(labels, values, color="darkgray")
plt.title('Bar Grapth')
```

✅ 4단계: 파이 차트 (Pie Chart)

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # 원을 동그랗게

```

✅ 5단계: 산점도 (Scatter Plot) & 스타일 커스터마이징
목표: 데이터 분포 표현 + 스타일 적용

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.title("산점도 예제")
plt.xlabel("X 값")
plt.ylabel("Y 값")
plt.colorbar()
plt.show()
```

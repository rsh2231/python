üßæ 1. Îã®Ïñ¥ Í∞úÏàò ÏÑ∏Í∏∞ (Î¨∏ÏûêÏó¥ Ï≤òÎ¶¨ + ÎîïÏÖîÎÑàÎ¶¨)
ÏÇ¨Ïö©ÏûêÎ°úÎ∂ÄÌÑ∞ Î¨∏Ïû•ÏùÑ ÏûÖÎ†•Î∞õÏïÑ Í∞Å Îã®Ïñ¥Í∞Ä Î™á Î≤à ÎÇòÏôîÎäîÏßÄ ÏÑ∏Ïñ¥ Ï∂úÎ†•ÌïòÏÑ∏Ïöî.

ÏûÖÎ†• ÏòàÏãú:I love python because python is fun
Ï∂úÎ†• ÏòàÏãú:
I: 1  
love: 1  
python: 2  
because: 1  
is: 1  
fun: 1


```python
str = input("Î¨∏Ïû•ÏùÑ ÏûÖÎ†•ÌïòÏÑ∏Ïöî:")
lst = str.split()

word_count = {}
for word in lst:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
        
for word, count in word_count.items():
    print(f"{word}: {count}")
    
```

üßë‚Äçüç≥ 2. ÏùåÏãù Ï£ºÎ¨∏ ÏãúÏä§ÌÖú (Ï°∞Í±¥Î¨∏ + Î∞òÎ≥µÎ¨∏ + ÎîïÏÖîÎÑàÎ¶¨)

ÏòàÏãú Î©îÎâ¥:

```python
menu = {"ÍπÄÎ∞•": 3000, "ÎùºÎ©¥": 3500, "Îñ°Î≥∂Ïù¥": 4000}

while True:
   order = input("ÏùåÏãùÏùÑ Ï£ºÎ¨∏ÌïòÏÑ∏Ïöî (Ï¢ÖÎ£åÌïòÎ†§Î©¥ 'Ï¢ÖÎ£å' ÏûÖÎ†•)")
   if order == 'Ï¢ÖÎ£å':
       print("Ï£ºÎ¨∏ÏùÑ Ï¢ÖÎ£åÌï©ÎãàÎã§.")
       break
   elif order in menu:
       print(f"{order}Ïùò Í∞ÄÍ≤©ÏùÄ {menu[order]}ÏûÖÎãàÎã§.")
   else:
       print("Ï£ºÎ¨∏ÌïòÏã† Î©îÎâ¥Í∞Ä ÏóÜÏäµÎãàÎã§.")
```
